```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('historico-nombres.csv', nrows = 5432786)
```

```python
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
plt.figure(figsize = (20,8))
plt.ylabel('# de nombres')
df.groupby('anio').count()['nombre'].plot(kind = 'bar')
plt.tight_layout()

plt.savefig('num_nombres_por_anio.pdf', bbox_inches='tight')
```

```python
plt.figure(figsize = (20,8))
plt.ylabel('# de personas')
df.groupby('anio').sum()['cantidad'].plot(kind = 'bar')
plt.tight_layout()

plt.savefig('num_personas_por_anio.pdf', bbox_inches='tight')
```

```python
df = df[df['nombre'].notnull()]

df['# de palabras'] = df['nombre'].apply(lambda x: len(x.split()))
```
